### TrainTest

```python
from sklearn.model_selection import train_test_split
```

```python
trainset, testset = train_test_split(df, test_size=0.2, random_state=0)
```

### Encodage

```python
from sklearn.impute import SimpleImputer

def impute_all(df, threshold=0.8):
    # Drop colonnes trop manquantes
    missing_rate = df.isna().mean()
    df = df.drop(columns=missing_rate[missing_rate > threshold].index)

    # Imputer les colonnes num√©riques avec la m√©diane
    num_cols = df.select_dtypes(include=['number']).columns
    num_imputer = SimpleImputer(strategy='median')
    df[num_cols] = num_imputer.fit_transform(df[num_cols])

    # Imputer les colonnes objets (cat√©gorielles) avec la valeur la plus fr√©quente
    cat_cols = df.select_dtypes(include=['object']).columns
    cat_imputer = SimpleImputer(strategy='most_frequent')
    df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])

    return df
    
```

```python
from sklearn.preprocessing import OneHotEncoder
def one_hot_encode(df):
    # S√©lection des colonnes cat√©gorielles
    cat_cols = df.select_dtypes(include=['object', 'category']).columns

    # Instanciation de l'encodeur
    encoder = OneHotEncoder(sparse_output=False, drop='first')  # drop='first' pour √©viter la multicolin√©arit√©

    # Transformation
    encoded = encoder.fit_transform(df[cat_cols])

    # R√©cup√©ration des noms de colonnes
    encoded_cols = encoder.get_feature_names_out(cat_cols)

    # Cr√©ation du DataFrame encod√©
    encoded_df = pd.DataFrame(encoded, columns=encoded_cols, index=df.index)

    # Suppression des colonnes d'origine et concat√©nation avec les nouvelles
    df = df.drop(columns=cat_cols)
    df = pd.concat([df, encoded_df], axis=1)

    return df
```

```python
from sklearn.preprocessing import StandardScaler

def apply_standard_scaler(df):
    # S√©lection des colonnes num√©riques
    num_cols = df.select_dtypes(include=['int64', 'float64']).columns

    # Instanciation du scaler
    scaler = StandardScaler()

    # Transformation
    df_scaled = df.copy()
    df_scaled[num_cols] = scaler.fit_transform(df[num_cols])

    return df_scaled
```

```python
def preprocessing(df):
    df = impute_all(df)
    df = one_hot_encode(df)
    df = apply_standard_scaler(df)

    X = df.drop('SalePrice', axis = 1)
    y = df['SalePrice']
    return X, y
```

```python
X_train, y_train = preprocessing(trainset)
```

```python
X_test, y_test = preprocessing(testset)
```

### Using pipeline

```python
df = Ames_housing.copy()
def dropnan(df, threshold=0.8):
    # Drop colonnes trop manquantes
    missing_rate = df.isna().mean()
    df = df.drop(columns=missing_rate[missing_rate > threshold].index)
    return df
df = dropnan(df)    
df.shape
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipeline = make_pipeline(SimpleImputer(strategy="median"),
StandardScaler())
cat_pipeline = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore"))
```

```python
from sklearn.compose import make_column_selector, make_column_transformer
preprocessing = make_column_transformer(
    (num_pipeline, make_column_selector(dtype_include=np.number)),
    (cat_pipeline, make_column_selector(dtype_include=object)),
)
```

```python
df = preprocessing.fit_transform(df)
```

```python
train_set, test_set = train_test_split(df,test_size=0.2, random_state=0)
```
